```python
import pandas as pd
import numpy as np
```

```python
#Reading in the data
df = pd.read_csv('DatascienceJobs.csv')
```

# Data cleaning

```python
df.head()
```

```python
df.columns
```

```python
#dropping irrelevant columns
df.drop(columns=['Unnamed: 0'], inplace=True, axis=1)
```

```python
df.head(3)
```

```python
#renaming of column names
df.rename({'Company Name': 'company_name'}, inplace=True , axis=1)
df.rename({'Type of ownership':'Type_of_ownership'}, inplace=True, axis=1)
df.rename({'Salary Estimate': 'Salary_estimate'}, inplace=True, axis=1)
df.rename({'Job Title': 'Job_title'}, inplace=True, axis=1)
df.rename({'Job Description': 'Job_Description'}, inplace=True, axis=1)

df.columns
```

```python
#removing unwanted strings in the company_name columns
df['company_name'] = df['company_name'].str.replace(r'(\r\n\d\.\d)', '')
df.head()
```

```python
#removing salary estimate having -1
df = df[df['Salary_estimate'] != '-1']
# df = df[df['Competitors'] != '-1']
df.shape
```

```python
#working on the salary estimate columns
df['Salary_estimate'] = df['Salary_estimate'].apply(lambda x: x.split('(')[0])
df['Salary_estimate'] = df['Salary_estimate'].apply(lambda x: x.replace('K', '').replace('$', ''))
df['hourly'] = df['Salary_estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0) 
df['employer_provided'] = df['Salary_estimate'].apply(lambda x: 1 if 'Employer Provided Salary' in x else 0)
#removing the employer provided salary and the hourly in the salary estimate columns
df.Salary_estimate = df['Salary_estimate'].apply(lambda x: x.lower().replace('per hour', '').replace('Employer Provided Salary:', ''))
```

```python
#splitting the salary estimate into min and max range columns

df['min_salary'] = df['Salary_estimate'].apply(lambda x: x.split('-')[0])
df['max_salary'] = df['Salary_estimate'].apply(lambda x: x.split('-')[1])
df['min_salary'] = df['min_salary'].str.replace('employer provided salary:', '')
# max_salary
df.head(100)
```

```python
#converting the new columns (min and max salary) to a type int
df.min_salary = df.min_salary.astype('int64')
df.max_salary =df.max_salary.astype('int64') 
df.dtypes
```

```python
#creating average salary columns
df['average_salary'] = (df.min_salary+df.max_salary)/2
df
```

```python
#Creating a job state column
df['job_state'] = df.Location.apply(lambda x: x.split(',')[1])
df[0:2]
```

```python
df['company_age'] = df['Founded'].apply(lambda x: 2022 - x if x > 1 else 0)
df[0:2]
```

```python
#same state
df['same_state'] = df.apply(lambda x: 1 if x.Location == x.Headquarters else 0, axis=1)
df[:4]
```

```python
#parsing the job description
'''important data science tools we should look out for in the job description column
-python
-r studio
-apache spark
-sql
-aws
-matlab
-tableau
'''
#python
df['python'] = df['Job_Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
# df['python'].value_counts()
'''------'''
#sql
df['SQL'] = df['Job_Description'].apply(lambda x: 1 if 'sql' in x.lower() else 0)
# sql.value_counts()
'''------'''
#spark
df['Spark'] = df['Job_Description'].apply(lambda x : 1 if 'spark' in x.lower() else 0)
# spark.value_counts()
'''------'''
#Aws
df['AWS'] = df['Job_Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
# aws.value_counts()
'''------'''
# power bi
df['POWERBI'] = df['Job_Description'].apply(lambda x: 1 if 'powerbi' in x.lower() else 0)
# power_bi.value_counts()
'''------'''
#excel
df['Excel'] = df['Job_Description'].apply(lambda x : 1 if 'excel' in x.lower() else 0)
# excel.value_counts()
'''------'''
#rstudio
df['R_studio'] = df['Job_Description'].apply(lambda x: 1 if 'r studio' in x.lower() else 0)
# r_studio.value_counts()
'''------'''
#matlab
df['matlab'] = df['Job_Description'].apply(lambda x: 1 if 'matlab' in x.lower() else 0)
# matlab.value_counts()
'''------'''
df['Tableau'] = df['Job_Description'].apply(lambda x: 1 if 'tableau' in  x.lower() else 0)

df.head()
# df[['python', 'SQL', 'Spark', 'AWS',  'POWERBI', 'Excel', 'R_studio', 'matlab', 'Tableau']].value_counts()
```

```python
#exporting the file to a csv format
df.to_csv('DataScienceJobs_cleaned.csv')
```
